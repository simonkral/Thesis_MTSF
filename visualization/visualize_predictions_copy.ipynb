```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python

def npy_to_csv(npy_path, csv_path, feature_names=None):
    data = np.load(npy_path)
    num_samples, pred_len, num_features = data.shape
    data_2d = data.reshape(num_samples * pred_len, num_features)

    if feature_names is None:
        feature_names = [f'feature_{i}' for i in range(num_features)]

    df = pd.DataFrame(data_2d, columns=feature_names)
    df.to_csv(csv_path, index=False)
    print(f"Saved CSV to: {csv_path}")
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- Helper functions ---
def npy_to_csv(npy_path, csv_path):
    data = np.load(npy_path)
    print(f"Loaded {npy_path} with shape {data.shape}")
    if data.ndim == 3:
        data = data[-1]  # Use last sample
    df = pd.DataFrame(data, columns=[f"feature_{i}" for i in range(data.shape[1])])
    df.to_csv(csv_path, index=False)

def load_last_steps(csv_path, steps):
    with open(csv_path) as f:
        total_rows = sum(1 for _ in f) - 1  # subtract header
    skip = max(total_rows - steps, 0)
    return pd.read_csv(csv_path, skiprows=range(1, skip + 1))

# --- Model prediction data ---
path = "_weather_shower_lag_720_T_degC_336_720_Linear_final_custom_ftM_sl336_ll48_pl720_dm512_nh8_el2_dl1_df2048_fc1_ebtimeF_dtTrue_rs2021_chCD_cdwd0.0_Exp_0"
base_dir = "/pfs/work9/workspace/scratch/ma_skral-SK_thesis_2025/Thesis_MTSF/PatchTST/PatchTST_supervised/results/"
npy_paths = [f"{base_dir}{path}/true.npy", f"{base_dir}{path}/pred.npy"]
csv_paths = [p.replace(".npy", ".csv") for p in npy_paths]

for npy_path, csv_path in zip(npy_paths, csv_paths):
    if not os.path.exists(csv_path):
        npy_to_csv(npy_path, csv_path)

df_true = load_last_steps(csv_paths[0], steps=720)
df_pred = load_last_steps(csv_paths[1], steps=720)

# --- Raw data ---
raw_df = pd.read_csv("/pfs/work9/workspace/scratch/ma_skral-SK_thesis_2025/Thesis_MTSF/PatchTST/PatchTST_supervised/dataset/_weather_shower_lag_720_T_degC.csv")
raw_df = raw_df[:-12]
raw_df = raw_df.iloc[-1440:].reset_index(drop=True)
raw_df.index = raw_df.index - 720  # shift so prediction starts at 0

# --- Plotting ---
fig, axes = plt.subplots(1, 2, figsize=(10, 5), sharey=False)

# --- Subplot 1: Predictions vs True ---
for channel in [1, 2]:
    axes[0].plot(df_pred[f"feature_{channel-1}"], label=f"pred_ch.{channel}")
    axes[0].plot(df_true[f"feature_{channel-1}"], label=f"true_ch.{channel}")
axes[0].set_title("Predictions vs True Values")
axes[0].set_xlabel("Time Steps")
axes[0].set_ylabel("Values")
axes[0].grid()

# --- Subplot 2: Raw Data with Lookback Windows ---
axes[1].plot(raw_df["T_degC"], label='true_ch.1')
axes[1].plot(raw_df["T_degC_LAG"], label='true_ch.2')
axes[1].axvline(x=0, color='black', linestyle='--', label='Start prediction, 720 ahead')
axes[1].axvline(x=(720-336+1-721), color='red', linestyle='--', label='Start lookback')
axes[1].axvline(x=(1440-336+1-721), color='red', linestyle='--', label='End prediction - lookback')
axes[1].set_title("Raw Data & Prediction Window")
axes[1].set_xlabel("Time Steps")
axes[1].grid()



# --- Shared Legend ---
handles, labels = axes[0].get_legend_handles_labels()
handles2, labels2 = axes[1].get_legend_handles_labels()
fig.legend(handles + handles2, labels + labels2, loc='lower center', ncol=4, bbox_to_anchor=(0.5, -0.01))

plt.tight_layout(rect=[0, 0.12, 1, 1])  # leave space for legend
os.makedirs("plots/Predictions", exist_ok=True)
plt.savefig("plots/Predictions/weather-lag720_ch-1_2_Linear-CD+Raw.png", bbox_inches='tight')
plt.show()
```
